#groupBy

```python
data = {'Company': ['GOOG','GOOG','MSFT','MSFT','FB','FB'], 'person':['Sam','Charlie','Amy', 'Vanessa','Carl','Sarah'],'Sales':[200,120,340,124,243,350]}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
byComp = df.groupby('Company')
```

```python
byComp.mean(numeric_only=True)
```

```python
byComp.sum(numeric_only=True)
```

```python
byComp.mean(numeric_only=True).loc['FB']
```

```python
df.groupby('Company').count()
```

```python
df
```

```python
df.groupby('Company').max()
```

```python
df.groupby('Company').min()
```

```python
df.groupby('Company').describe() #full overview  if we wanna know , this groupby much more important for exam
```

```python
df.groupby('Company').describe().loc['MSFT']
```

```python
df.groupby('Company').describe().transpose()
```
